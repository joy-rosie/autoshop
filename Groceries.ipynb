```python
groceries = auto_shop.get_groceries()
```

```python
xpath_product_input_amount = "//input[@type='number']"
xpath_add = "//span[text()='Add']"
xpath_checkout_to_confirm_changes = "//span[text()='Checkout to confirm changes']"

for row in groceries["values"][1:]:
    name = row[1]
    url = row[2]
    amount = row[3]
    description = row[4]
    not_required = row[5].lower() == "true"
    
    info = f"{name=}, {url=}, {amount=}, {description=}, {not_required=}"

    if not_required:
        auto_shop.logger.debug(f"Not required - {info=}")
    else:
        try:
            driver.get(url)
            auto_shop.wait_and_send_keys_and_delete(driver=driver, value=xpath_product_input_amount, keys=amount)
            auto_shop.wait_and_click(driver=driver, value=xpath_add)
            _ = auto_shop.wait_and_get(driver=driver, value=xpath_checkout_to_confirm_changes)
        except KeyboardInterrupt as e:
            raise e
        except Exception as exception:
            auto_shop.logger.error(f"Failed - {info=}, {exception=}")
```
